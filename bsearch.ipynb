```python
# Returns index of x in arr if present, else return -1
def binary_search(arr, low, high, x):
 
    
    if high >= low:
 
        mid = (high + low) // 2 # taking mid position
 
        # If element is present at the middle itself
        if arr[mid] == x:
            return mid
 
        # If element is smaller than mid, then it can only it should be at left of mid element
        
        elif arr[mid] > x:
            return binary_search(arr, low, mid - 1, x)#array,left and right indexes,element to search
 
        #array,left and right indexes,element to search
        else:
            return binary_search(arr, mid + 1, high, x)
 
    else:

        return -1         # Element is not present 
 
# Test array
arr = [ 2, 8, 15, 35, 40 ]
#inf the array is not sorted order do that first
x = 35
 
# calling function for search
result = binary_search(arr, 0, len(arr)-1, x) #left most and right most element's index values , result returns the index value
 
if result != -1:
    print("Element is present at position", str(result+1))
else:
    print("Element is not present in array")
```
